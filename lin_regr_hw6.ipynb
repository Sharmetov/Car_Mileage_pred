```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
import pandas as pd

data = pd.read_csv('carmpg.csv')
mpg = np.array(data['MPG'])
hp = np.array(data['HP'])
# print(hp)
```

```python
x_mean = np.mean(hp)
y_mean = np.mean(mpg)

slope = (np.sum((hp-x_mean)*(mpg-y_mean)))/np.sum(np.square(hp-x_mean))
print(slope)

intercept = y_mean - slope * x_mean
print(intercept)
```

```python
y = intercept + slope*hp
```

```python
# Clarify with the built in functions 
slope, intercept, r, p, std_err = stats.linregress(hp, mpg)
print(slope)
print(intercept)
```

```python
plt.scatter(hp, mpg)
plt.xlabel('HP')
plt.ylabel('MPG')
plt.plot(hp, y, color='red', label = f'Best Fit line: y = {slope:.2f}x + {intercept:.2f}')
plt.legend()
plt.show()
```
